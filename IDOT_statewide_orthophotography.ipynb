# 2011 Illinois Department of Transportation (IDOT) Orthophotography

[IDOT created a state-wide collection of orthophotograpy](https://clearinghouse.isgs.illinois.edu/data/imagery/2011-illinois-department-transportation-idot-orthophotography)
consisting of photographic images covering the entire state. As with with the proposed DEM collection, IDOT faced the challenge of which projection to use for 
the images. Rather than use a nation-wide projection such as the [U.S. National Atlas Equal Area](http://spatialreference.org/ref/epsg/us-national-atlas-equal-area/)
or an uncommon projection, IDOT choose to use State Plane projection and divide the state vertically. Eastern counties use the NAD83 / Illinois East (ftUS) ([EPSG:3435](http://spatialreference.org/ref/epsg/3435/)) and the western counties use NAD83 / Illinois West (ftUS) ([EPSG:3436](http://spatialreference.org/ref/epsg/3436/)). 
To simplify processing some duplication of tiles is allowed near the east-west divide. The tiling based on the two projection system can be seen below. 

```python
%matplotlib inline
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
il = gpd.read_file('../examples/IL_Counties.gpkg')
print('IL =', len(il), il.crs)
SPE = gpd.read_file('../examples/maps/SPE_10000.shp')
print('SPE =', len(SPE), SPE.crs)
SPW = gpd.read_file('../examples/maps/SPW_10000.shp')
print('SPW =', len(SPW), SPW.crs)

fig, ax = plt.subplots(figsize=(10,30))
SPE4326 = SPE.to_crs(il.crs)
SPE4326.plot(ax=ax, facecolor='none', edgecolor='black')
SPW4326 = SPW.to_crs(il.crs)
SPW4326.plot(ax=ax, facecolor='none', edgecolor='black')
il.plot(ax=ax, facecolor='none', edgecolor='red')
plt.show()
```
