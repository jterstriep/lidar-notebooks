```python
import os
import geopandas as gpd

lidar = '../lidar'
counties = sorted(os.listdir(lidar))
print(counties)
```

```python
for c in counties:
    dbase = os.path.join(lidar, c, 'tiles.gpkg')
    if os.path.exists(dbase):
        tindex = gpd.read_file(dbase)
        print(c, tindex.total_bounds, tindex['srs'].unique())
```
